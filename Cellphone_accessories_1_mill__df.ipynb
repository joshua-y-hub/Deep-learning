```python
import os
import pandas as pd

# Update the file path to point to your Google Drive
file_path = "/content/drive/MyDrive/Cell_Phones_and_Accessories_5.json"

if os.path.exists(file_path):
    print(f"The file '{file_path}' exists. Proceeding to load it.")
    df = pd.read_json(file_path, lines=True)
    display(df.head())
else:
    print(f"Error: The file '{file_path}' does not exist in your Google Drive. Please ensure the file is in the root of 'My Drive' or adjust the path accordingly.")
```

```python
df= df.drop(['vote', 'image', 'verified'], axis=1).copy()
df.head()
```

```python
df.shape
```

```python
# Filter out NaN values before applying text preprocessing
review_text = df.reviewText.dropna().apply(gensim.utils.simple_preprocess)

# Alternative approach: Fill NaN values with empty strings first
# review_text = df.reviewText.fillna('').apply(gensim.utils.simple_preprocess)
```

```python
review_text
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model_save_path = "/content/drive/MyDrive/word2vec-amazon-cell-accessories-reviews-short-GPU.model"
model.save(model_save_path)

print(f"Model saved successfully to: {model_save_path}")
```
